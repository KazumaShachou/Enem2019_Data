## Machine Learning

```
#we will entry note and  exit note to try to predict this notes of math
#but first we need to remove every "NAN" from nozero, cause some note of enem have still NAN, so we will to use a drop.na

nozero = nozero[tests].dropna()

entry = ['NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_CN', 'NU_NOTA_REDACAO']
exit = 'NU_NOTA_MT'

#now we will to create a datafram with this entry note and exist note using another main dataframe (nozero)

entry_note = nozero[entry]
exit_note = nozero[exit]
```

```
 entry_note
```

```
#it's like a f(x) = y , a function  , so x it's like a entry note and y it's like a exit note
x = entry_note
y = exit_note
```

```
#now to create a model, to try to predict of 1 person that do the enem
#first we will start to separe the volume of data and another one to make a test of our model 
#lets divide training data, and data test
# we will to use scikitlearn to make this

from sklearn.model_selection import train_test_split

train_test_split(x, y)
```

```
 #this framework will return to us x_train, x_test, y_train, y_test = train_test_split(x,y) , now let's use test_size
 #we need to create a seed, for use to avoid random results just using "test_size", because this one has a random factor

 seed = 9876

 x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = seed) 
```

```
#let's do our 1st model, we will use svm(suport vector machine), this make a linear model
from sklearn.svm import LinearSVR

#this linear SVR have a random value too, so we need to use a random_state
model = LinearSVR(random_state = seed)
#now we need to caught our date and make a fit
model.fit(x_train, y_train)
```

```
#now we will to use the "entry" command, with that subjects and return the prediction of math
#this will return a array with math grades
predict_math = model.predict(x_test) 
predict_math
```

```
#lets build a metric that can evaluate our datas
results = pd.DataFrame()

results['Real'] = y_test
results['Forecast'] = predict_math

#we can make the difference between the data true and the column test
#we can avoid some type of noise in analyze , just use the square number (so we will transform an error 100 in 10k and -100 in 10k for example, because is more easy to understand that is a error)
results['Diference'] = results['Real'] - results['Forecast']
results['Diference_persquare'] = (results['Real'] - results['Forecast'])**2 
```

```
#we will use a type of model that to separe in 5 pieces an study object, for example the subjetcs, so we use 20% of them, or 1/5 in folds
# the first one is a test and the 4/5 are the  train , so we train 80% and test 20%
#second test we use the fold 2, the second piece (20%) and another ones to train, in 3rd test we use the  3 piece(fold) to test and another ones train , until the 5 piece
#when we got 5 results, we just use this 5 results and sum and for last take the mean , it's a good model

from sklearn.model_selection import cross_validate

#we will use the scording , it's a standard that we can choose that we want to work, so we will work with mean squared error, inside a string, that we worked in another models
#but for we use our mean squared error, in documentation of crossvalidation , we need to use "neg"
#now let's use our tree model
tree_model = DecisionTreeRegressor(max_depth= 2 )
#now we will use the crossvalidate to aply in our trained model
results = cross_validate(tree_model, x, y, scoring = "neg_mean_squared_error")
#if I want to return this arrays positives, we just need to multiply for minus 1
results['test_score']* -1
```

```
  #Crossvalidate return some things, like fit time, score time and test score
#test score is a variation of each folders that was trained, he used automatically 5 folds, but if we want we can change too
#in sktlearn have a rule that how much near of 0 , better, cause your accuracy is better

#now the mean of results
mean = (results['test_score']*-1).mean()  
mean
```

```
#if I want to use more models to train, just need to use cv in results, let's use 30 now
results = cross_validate(tree_model, x, y, scoring='neg_mean_squared_error', cv= 30)
results
```

```
#we need to calculate standard deviation to know the accuracy of our model

standard_deviation  = (results['test_score']* -1 ).std()
standard_deviation
```

```
#we wil to calculate the confidence interval to more acc , so we need a limit inferior and lim upper
lim_inferior =  mean - (2*standard_deviation)
lim_upper = mean + (2*standard_deviation)

print(f'confidence interval{lim_inferior} - {lim_upper}')
```

```
 #this result mean that 95% of prevision is inside of this stantard 6162 to 6950
 
 #let's make a function that will print some values without need to rewrite again everything using def

def calculate_mse(results):
  mean = (results['test_score']*-1).mean()  
  standard_deviation  = (results['test_score']* -1 ).std()
  lim_inferior =  mean - (2*standard_deviation)
  lim_upper = mean + (2*standard_deviation)

  print(f'confidence interval {lim_inferior} - {lim_upper}')
```

```
calculate_mse(results)
```

```
#lets import kfold , now we could to divide our datas
#so we need to use suffle for mix our results 
#we need to import numpy for use a seed , so we will not change the results when enter

import numpy as np

Seed =  9999
np.random.seed(Seed)

from sklearn.model_selection import cross_validate
from sklearn.model_selection import KFold


pieces = KFold(n_splits = 10, shuffle= True)
tree_model = DecisionTreeRegressor(max_depth= 5 )
results = cross_validate(tree_model, x, y, scoring='neg_mean_squared_error', cv= pieces)
calculate_mse(results)
```

```
def regressor_tree(nivel):
    SEED= 1232
    np.random.seed(SEED)
    parts = KFold(n_splits = 10, shuffle=True)
    tree_model = DecisionTreeRegressor(max_depth=nivel)
    results = cross_validate(tree_model, x, y, cv= parts, scoring="neg_mean_squared_error", return_train_score=True)
    print(f"Train = {(results['train_score']*-1).mean()}|Test = {(results['test_score']*-1).mean()}")
```

```
#let's use "for i" and do the level of deeph to very to 1 until 20
for i in range(1, 21):
  regressor_tree(i)
```

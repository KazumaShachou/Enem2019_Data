## exploratory analysis

```
#lets search and find the state of youngs 13-14 years people that had this test 
enem.query('NU_IDADE <= 14')['SG_UF_RESIDENCIA'].value_counts()
```

```
#proportion of participation of each state
enem.query('NU_IDADE <= 14')['SG_UF_RESIDENCIA'].value_counts(normalize = True)
```

```
#not recomended to use pie graphics with many pieces, because can be confuse and not worth
less15yo_students = enem.query('NU_IDADE < 15')
less15yo_students['SG_UF_RESIDENCIA'].value_counts().plot.pie(figsize=(9,9));
```

```
#lets use the bar graph, is more easy to visualize 
less15yo_students['SG_UF_RESIDENCIA'].value_counts(normalize = True).plot.bar(figsize = (12, 6));
```

```
#so now let's use the data to know about the target age, or the most people that enter use this test to enter in college (17-20years)
target_age = enem.query('NU_IDADE <= 20')
target_age['SG_UF_RESIDENCIA'].value_counts(normalize = True).plot.bar(figsize = (12,6));
```

```
#you can observe this number represent a trustworthy result of the number os people that make the exame in Brazil per state in this data
#for prove this, we will see in the same graph , people until the oldest age in graph , we will observe not much change.
geral = enem.query('NU_IDADE <= 82') 
geral= enem['SG_UF_RESIDENCIA'].value_counts(normalize = True).plot.bar(title = 'All Ages',  figsize = (12, 6))
```

```
#now let's get the question 6 of score from Nota_LC and use a boxplot to show the results
#Q006 = whice is the your family income ? add  your income with your family and divide per number os members
#we have the A for minor familiy income, and Q for the biggest family income (A-Q)

#We will to use seaborn for boxplot, now we just need to call seaborn with boxplot and tell the x is the question 6 and y is a score of school subject
#the end we just need to put an index and organize the A until Q
#the "data" inside boxplot are from the data where we will to use 

import seaborn as sns
import matplotlib.pyplot as plt

income = enem['Q006'].unique()
income.sort()

plt.figure(figsize=(10,6))  
sns.boxplot(x= 'Q006', y = 'NU_NOTA_MT', data = enem, order= income);
plt.title('Boxplot of MT');
```

```
#in the graph we observed that more lower is the family income, less score they had, the outlayer decrease too(Black points in graph that represent 1.5%)
#now let's see the LC score per income
#Here we see the same thing where the median is higher per family income, but we can oberve the outlayer increasing when we see the lower scores per income family
#It's happened because the same reason of the past graph, this outlayer represent 1.5% of the people that had lower scores, but for lower income family people not have this outlayer, because 24% of the people had the lowest score
plt.figure(figsize=(10,6))  
sns.boxplot(x= 'Q006', y = 'NU_NOTA_LC', data = enem, order= income);
plt.title('Boxplot of LC');
```

```
#now let's sum the scores per line in axis 1
enem[tests].sum(axis = 1)
```

```
#now for add 1 more column with total score called NOTA_TOTAL
enem['NU_NOTA_TOTAL'] = enem[tests].sum(axis = 1)
enem.head(10)
```

```

#now let's use the boxplot graph using "NOTA_TOTAL" for compare with the family income
plt.figure(figsize=(18,7))
sns.boxplot(x = 'Q006', y = 'NU_NOTA_TOTAL', data = enem, order = income) 
plt.title('TOTAL DE NOTAS');
```

```
#now let's see what is happening, why so many 0, we can see that maybe is something wrong 

sns.displot(enem, x = 'NU_NOTA_TOTAL');
```

```
#the will to use an append in 'NU_NOTA_TOTAL for just return the people that had 0 in enem
#The "NAN" returned in scores can be explained when the person do the test, they put a "NAN" , not a 0.

tests.append('NU_NOTA_TOTAL')
enem[tests].query('NU_NOTA_TOTAL == 0')
```

```
#here we get the people that come at least 1 day in exam
nozero = enem[tests].query('NU_NOTA_TOTAL != 0')
nozero.head()
```

```
plt.figure(figsize=(10,6))  
sns.boxplot(x= 'Q006', y = 'NU_NOTA_TOTAL', data = nozero, order= income);
plt.title('Boxplot of Total de notas');
```
